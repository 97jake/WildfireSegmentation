# Tutorial on how to find and download satellite imagery

This short tutorial will show you how to download your own satellite imagery and how to load it into python. If you get stuck somewhere, you can message @jalen morgan on slack.



### (a) Set up an account

To gather satellite imagery, you need an account on https://scihub.copernicus.eu/dhus/#/home. You can create an account by clicking the search button, then clicking 'Sign up.' Or you can go directly to https://scihub.copernicus.eu/dhus/#/self-registration. Once you have an account set up, go back to https://scihub.copernicus.eu/dhus/#/home and you can start specifying what images you're looking for.


### (b) Setting up the search

Next, specify some parameters to filter down your search. Click the funnel/three bars on the left hand side. Specify the "Sensing Period". This is when the image was taken (as opposed to the "ingestion period" which is when the image was uploaded to the database)


### (c) Select a type of satellite image

You'll see three sections below that identify what type of images you want. The three options are sentinel-1, sentinel-2, and sentinel-3. 

- Sentinel1 images contain one band of data. They are older, and the file sizes are a lot larger (About 6GB on average). Because they are so large and only contain one band of data, we won't be using these here. Although their resolution can be quite good
- Sentinel2 images contain 12 bands of data. They have good resolution at 33-200 ft per pixel (Depending on the band). The file sizes are 1GB on average
- Sentinel3 images contain 21 bands of data. They are the newest ones and are only available after 2016. They have a resolution of 980 ft (per pixel). File sizes are on average 500MB

__Sentinel2__  
Sentinel-2 is your best option if you want to look closer up at land. The resolution is ~33ft per pixel, which should get you an overview of the land. One downside of Sentinel2 is that the files are quite large. For Sentinel2, you can also specify cloud coverage, which can be useful when . Below are the bands given for Sentinel-2.


**Sentinel3**  
Sentinel-3 is a more zoomed out image, with more bands. One plus is the file sizes are about half the size of Sentinel2 images.

To Select a type of image, click on the checkbox on the left hand side. If choosing Sentinel-2, specify cloud cover % by using square brackets, [0 TO 10]


### (d) Select an area to image

Finally, specify an area to look at. You can select on the right hand side the highlight tool and highlight the area you'd like to get images for. Once the area is highlighted, you're ready to hit the search button. It may take a few minutes to locate the images you're looking for. There also might not be any images. If this is the case, you need to expand your parameters.


### (e) View and download results

 Assuming your search was successful, you'll get a list of results on the left. If you hover your mouse over one of the results, it'll show you the outline of that result. You probably want the outline to contain the area you highlighted. If it does, you're good to go. Hover over the image and click download.
 

If you're searching for old stuff, you'll run into the problem of images being "offline". This means that the images are old enough that they've moved them into long term storage. All you have to do is request to download the image, and within 30 minutes to an hour, they'll move that image to normal storage, and the "Offline" label should disappear.
